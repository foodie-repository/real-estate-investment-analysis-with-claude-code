주소 파싱, 면적대/평형대/연식 분류, 좌표 조인을 검증합니다.

```python
import sys
from pathlib import Path

# 프로젝트 루트를 sys.path에 추가
PROJECT_ROOT = str(Path.cwd().resolve().parents[2])
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

from src.utils.address import parse_sigungu, build_road_address_key
from projects.part09_trade_map.dashboard.preprocessing import get_region_options, get_year_range
```

### 1. 주소 파싱 검증

```python
# 다양한 시군구 형식 테스트
test_cases = [
    "서울특별시 강남구 압구정동",
    "경기도 성남시 중원구 은행동",
    "충청남도 예산군 예산읍 예산리",
    "충청남도 천안시 서북구 성환읍 율금리",
    "세종특별자치시  조치원읍 상리",  # 더블스페이스
    "세종특별자치시 세종시 한솔동",
    "부산광역시 서구",
]

for case in test_cases:
    result = parse_sigungu(case)
    print(f"입력: {case!r}")
    print(f"  → 시도={result['시도']}, 시군구={result['시군구']}, 읍면동={result['읍면동']}")
    print()
```

### 2. 도로명주소_key 생성 검증

```python
# 도로명주소 키 생성 테스트
test_keys = [
    ("서울특별시 강남구 압구정동", "압구정로 201"),
    ("경기도 성남시 중원구 은행동", "수내로192번길 25"),
    ("충청남도 예산군 예산읍 예산리", "아리랑로144번길 19"),
    ("세종특별자치시  조치원읍 상리", "세종로 1"),
]

for sigungu, road in test_keys:
    key = build_road_address_key(sigungu, road)
    print(f"시군구={sigungu!r}, 도로명={road!r}")
    print(f"  → 도로명주소_key = {key!r}")
    print()
```

### 3. 지역 옵션 데이터 확인

```python
# 시도 목록
regions = get_region_options()
print(f"시도 수: {len(regions)}")
for 시도 in sorted(regions.keys()):
    시군구_수 = len(regions[시도])
    print(f"  {시도}: {시군구_수}개 시군구")
```
